```python
def RollTheDice():

    import random

    diceNumbers = [1,2,3,4,5,6]
    
    diceRollNumber = 1

    keepRolling = True
    
    while keepRolling == True:
    
        dice1 = random.choice(diceNumbers)
        dice2 = random.choice(diceNumbers)
        
        print("Roll#", diceRollNumber, ":", dice1, dice2, dice1+dice2)
        
        if dice1 != dice2:
            keepRolling = False
        else:
            diceRollNumber = diceRollNumber + 1
            if diceRollNumber == 4:
                keepRolling = False
                print("You rolled doubles three times, off to Jail you go!")
    
    print("done")
        
RollTheDice()
```

```python
#Write a Python program that plays a game of rock, paper, scissors with the user. 
#Your program should prompt the user for their play, randomly generate its own, then announce a winner. 
#Your program should keep a running tally of the user’s wins and losses and ask if they want to play again, \
#starting a new game if they say “yes.”

def RockPaperScissors():
    
    import random

    rock_paper_scissors = ["rock","paper","scissors"]
    
    keepPlaying = input("Do you want to play rock, paper, scissors? (y/n)")
    print(" ")
    userTotalWins = 0
    userTotalLosses = 0
    userTies = 0
    
    while keepPlaying == "y":
    
        userChoice = input("Input rock, paper, or scissors: ")
        
        pgmChoice = random.choice(rock_paper_scissors)
        print("Program Choice: " + pgmChoice)
        
        if userChoice == "rock" and pgmChoice == "scissors":
            print("You: rock crushes me: Scissors.  You win")
            userTotalWins = userTotalWins + 1
        elif userChoice == "scissors" and pgmChoice == "rock":
            print("You: scissors got crushed by me: Rock.  You lose")
            userTotalLosses = userTotalLosses + 1
        elif userChoice == "paper" and pgmChoice == "rock":
            print("You: paper covers me: Rock.  You win")
            userTotalWins = userTotalWins + 1  
        elif userChoice == "rock" and pgmChoice == "paper":
            print("You: rock covered by me: Paper.  You lose")
            userTotalLosses = userTotalLosses + 1 
        elif userChoice == "scissors" and pgmChoice == "paper":
            print("You: scissors cuts me: paper.  You win")
            userTotalWins = userTotalWins + 1  
        elif userChoice == "paper" and pgmChoice == "scissors":
            print("You: paper cut by me: Scissors.  You lose")
            userTotalLosses = userTotalLosses + 1
        elif userChoice == "paper" and pgmChoice == "paper":
            print("We're tied")
            userTies = userTies + 1
        elif userChoice == "scissors" and pgmChoice == "scissors":
            print("We're tied")
            userTies = userTies + 1
        elif userChoice == "rock" and pgmChoice == "rock":
            print("We're tied")
            userTies = userTies + 1
        
        print("Your wins: ", userTotalWins, "Your losses: ", userTotalLosses, "Ties: ", userTies)
        print(" ")
        keepPlaying = input("Do you want to play rock, paper, scissors again? (y/n)")
        
   
    print("done")  

RockPaperScissors()
```
